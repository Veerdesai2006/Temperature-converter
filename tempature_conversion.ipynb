```python
#The temprature has mainly three unit kelvin, Degree celsius and fahrenheit
#so we are going to convert them form one to another using functions
def conversion_to_kelvin(value,unit):
    #c = celsius
    if unit=="c":
        #c = celsius
        k=value+273.15
        return k
    if unit=="f":
        #f = fahrenheit
        k=(value-32)*(5/9)+(273.15)
        return k
    if unit=="k":
        #k=kelvin
        print("Temprature is already in kelvin")
```

```python
def conversion_to_celsius(value,unit):
    if unit=="c":
        print("The temprature is already in celsius")
    if unit=="k":
        c=value-273.15
        return c
    if unit=="f":
      c=(value-32)*(5/9)
    return c
```

```python
def conversion_to_fahrenheit(value,unit):
    if unit=="c":
        f=(value*(9/5))+32
        return f
    if unit=="k":
        f=(value-273.15)*(9/5)+(32)
        return f
    if unit=="f":
        print("temprature is alreday into kelvin")    
```
